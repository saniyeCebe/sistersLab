```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns
```

```python
html_doc = "https://realpython.github.io/fake-jobs"
response = requests.get(html_doc) 
soup = BeautifulSoup(response.content, "html.parser")
```

```python

print(soup.prettify()) #BeautifulSoup ile ayrtrlm (parse edilmi) HTML ieriini gzel, okunabilir ve girintili (indentli) bir ekilde yazdrr.
```

```python
soup.find_all('a')
```

```python
for link in soup.find_all('a'):
    print(f"{link.get('id')} : {link.get('href')}")
```

```python
job_cards = soup.find_all('div', class_='card-content')

for card in job_cards:
    title = card.find('h2', class_='title').text.strip()
    company = card.find('h3', class_='company').text.strip()
    location = card.find('p', class_='location').text.strip()
    apply_link = card.find_all('a')[-1]['href']
    
    if "python" in title.lower():
        python_jobs.append({
            'title': title,
            'company': company,
            'location': location,
            'link': apply_link
        })
        
    
```

```python
print(f"\nPython ile ilgili toplam {len(python_jobs)} i ilan bulundu:\n")
for ilan in python_jobs:
    print(f"{ilan['title']} - {ilan['company']} ({ilan['location']})")
    print(f"Bavuru Linki: {ilan['link']}\n")
```

```python
# Lokasyonlara gre dalm
location_counts = Counter([job['location'] for job in python_jobs])

# Grselletirme
plt.figure(figsize=(10, 6))
sns.barplot(x=list(location_counts.values()), y=list(location_counts.keys()), palette='viridis')
plt.title('Python Geen  lanlarnn Lokasyona Gre Dalm')
plt.xlabel('lan Says')
plt.ylabel('Lokasyon')
plt.tight_layout()
plt.show()
```
